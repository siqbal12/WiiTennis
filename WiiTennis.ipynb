```python
import pandas as pd
import numpy as np
```

# Winning a Game

```python
df = pd.read_csv('Wii Tennis Stats Final Project - Probability of Winning a Game.csv').set_index('Unnamed: 0')
```

```python
df.iloc[-5, -5] = 0
df.iloc[-8, -7] = 0
```

```python
df
```

```python
t = df.values
t.shape
```

```python
q = t[:18, :18]
q.shape
```

```python
r = t[:18, 18:]
r.shape
```

```python
q
```

```python
i = np.identity(18)
i
```

```python
i - q
```

```python
n = np.linalg.inv(i-q)
n
```

```python
n.shape
```

```python
nr = n @ r
df_final = pd.DataFrame(nr)
df_final
```

# Winning a Set

```python
df = pd.read_csv('Wii Tennis Stats Final Project - Probability of Winning a Set.csv').set_index('Unnamed: 0').fillna(0)
```

```python
df
```

```python
t = df.values
t.shape
```

```python
q = t[:41, :41]
q.shape
```

```python
r = t[:41, 41:]
r.shape
```

```python
i = np.identity(41)
i.shape
```

```python
n = np.linalg.inv(i-q)
n.shape
```

```python
nr = n @ r
df_final = pd.DataFrame(nr)
df_final
```
